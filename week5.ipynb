```python
#task 1

import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("data.csv", sep=";", decimal=",")

res = [["регион", "р-уровень"]]
regions = df["регион"].unique()

for r in regions:
    group = df[df["регион"] == r]
    p_value = stats.mannwhitneyu(
        group["время_до"], group["время_после"], alternative="two-sided"
    ).pvalue

    res.append([r, float(p_value.round(4))])

res[1:] = sorted(res[1:], key=lambda x: x[1])

print(res)
```

```python
#task 2

import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("data.csv", sep=";", decimal=",")

res = [['показатель', 'р-уровень', 'метод расчёта']]

for ind in df.columns[1:]:
    gr1 = df[df['group'] == 'group_1'][ind]
    gr2 = df[df['group'] == 'group_2'][ind]

    t1 = stats.shapiro(gr1)[1] > 0.05
    t2 = stats.shapiro(gr2)[1] > 0.05

    if t1 and t2:
        p_value = stats.ttest_ind(gr1, gr2, equal_var=True).pvalue
        method = 'Стьюдент'
    else:
        p_value = stats.mannwhitneyu(gr1, gr2, alternative='two-sided').pvalue
        method = 'Манн-Уитни'

    res.append([ind, float(p_value.round(4)), method])

res[1:] = sorted(res[1:], key=lambda x: x[1])

print(res)
```

```python
#task 3

import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("data.csv", sep=";", decimal=",")

res = [['показатель', 'р-уровень']]

types = ['тип_осложнения_1', 'тип_осложнения_2',
         'тип_осложнения_3', 'тип_осложнения_4']

for t in types:
    contingency = pd.crosstab(df['вид_зондирования'], df[t])
    chi2, p_value, d, e = stats.chi2_contingency(contingency)

    res.append([t, float(p_value.round(4))])

res[1:] = sorted(res[1:], key=lambda x: x[1])

print(res)
```

```python
#task 4

import pandas as pd
import numpy as np
from scipy import stats
import itertools

df = pd.read_csv("data.csv", sep=";", decimal=",")

df1 = df[df['Группа'] == 'Группа_1']

res = [['показатель 1', 'показатель 2',
        'значение корреляции', 'р-уровень', 'метод корреляции']]

for (ind1, ind2) in itertools.combinations(df.columns[1:], 2):
    gr1 = df1[ind1]
    gr2 = df1[ind2]

    t1 = stats.shapiro(gr1)[1] > 0.05
    t2 = stats.shapiro(gr2)[1] > 0.05

    if t1 and t2:
        corr, p_value = stats.pearsonr(gr1, gr2)
        method = 'Пирсон'
    else:
        corr, p_value = stats.spearmanr(gr1, gr2)
        method = 'Спирмен'

    if p_value < 0.05:
        res.append([ind1, ind2, float(corr.round(2)),
                   float(p_value.round(4)), method])

res[1:] = sorted(res[1:], key=lambda x: x[2], reverse=True)

print(res)
```
