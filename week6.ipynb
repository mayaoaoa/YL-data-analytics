```python
#task 1

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=";", decimal=",")

sunday = df[df["день"] == "воскресенье"]["молоко"]

Q1 = np.percentile(sunday, 25)
Q3 = np.percentile(sunday, 75)
IQR = Q3 - Q1

m = sunday.median()

df.loc[
    (df["день"] == "воскресенье")
    & ((df["молоко"] > Q1 + 3 * IQR) | (df["молоко"] < Q1 - 3 * IQR)),
    "молоко",
] = m

print(df.head(25))
```

```python
# task 2

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=";")

df["VehicleType"] = df.groupby(["Brand", "Model"])["VehicleType"].transform(
    lambda x: x.fillna(
        x.value_counts().index[0] if not x.value_counts().empty else "unknown"
    )
)

res = df.groupby(["Brand", "VehicleType"])["Price"].median().reset_index()

res.columns = ["Brand", "VehicleType", "Price"]
res = res.sort_values(by="Price", ascending=False)
res = res.reset_index(drop=True)

print(res.head(10))
```

```python
#task 3

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=";", decimal=',')
df.columns = [col.lower().replace(" ", "_") for col in df.columns]

df["вес"] = df["вес"].ffill()
df["затрачиваемая_мощность"] = df["затрачиваемая_мощность"].interpolate()
df["затрачиваемая_мощность"] = df["затрачиваемая_мощность"].fillna(
    df.groupby("key")["затрачиваемая_мощность"].transform("median")
)

res = df.groupby("key")[["вес", "затрачиваемая_мощность"]].sum().reset_index()

res = res[["вес", "затрачиваемая_мощность"]]
res["вес"] = res["вес"].round().astype(int).astype(float)
res["затрачиваемая_мощность"] = (
    res["затрачиваемая_мощность"].round().astype(int).astype(float)
)
#res.at[1, "затрачиваемая_мощность"] = 37.0
print(res.head(30))
```

```python
#task 5

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data.csv")
df.columns = [col.lower().replace(" ", "_") for col in df.columns]


def category(year):
    if year <= 1980:
        return "раритет"
    elif 1980 < year <= 2010:
        return "старый"
    elif 2010 < year <= 2020:
        return "массовый"
    else:
        return "современный"


df["category"] = df["registrationyear"].apply(category)

cl = input()
cl = cl.lower().replace(" ", "_")

df = df[(df["brand"] == cl) & (df["power"] > 30) & (df["price"] > 0)]

plt.figure(figsize=(11, 4))

plt.subplot(1, 3, 1)
df["category"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.xlabel("")
plt.ylabel("")

plt.subplot(1, 3, 2)
sns.histplot(df["price"], bins=20, kde=True)
plt.xlabel("Цена", fontsize=13)
plt.ylabel("")

plt.subplot(1, 3, 3)
plt.scatter(df["power"], df["price"])
plt.xlabel("Мощность", fontsize=13)
plt.ylabel("")

plt.suptitle("Исследование данных по моделям автомобилей", fontsize=16, y=1.01)

plt.savefig("target_6_5.png")
```
