```python
#task 1

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=";")

n = int(df['число_опрошенных'][0])
p = float(df['доля_ответов'][0].replace(',', '.'))
z = float(df['доверительный_уровень'][0].replace(',', '.'))

H = z * np.sqrt((p * (1 - p)) / n)

print(round((p - H) * 100, 2), round((p + H) * 100, 2))
```

```python
#task 2

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=",")

n = 200
p = (df["пробег"] > 140).sum() / n
z = 1.96

H = z * np.sqrt((p * (1 - p)) / n)

print(round((p - H) * 100, 2), round((p + H) * 100, 2))
```

```python
#task 3

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=";", decimal=",")

df = df[df["название"] == "Полёт"]

Q1 = df["вес_шоколадки"].quantile(0.25)
Q3 = df["вес_шоколадки"].quantile(0.75)

IQR = Q3 - Q1

df = df[
    (df["вес_шоколадки"] >= Q1 - 1.5 * IQR) & (df["вес_шоколадки"] <= Q3 + 1.5 * IQR)
]

z = 1.96
H = 1 

n = (z * df['вес_шоколадки'].var() / H) ** 2

print(round(n, 0))
```

```python
#task 4

import pandas as pd
import numpy as np

H = np.arange(0.01, 0.10, 0.01)
res = []

z = 1.96
p = 0.5

i = 1

for h in H:
    n = (z ** 2 * p * (1 - p)) / (h ** 2)
    res.append((i, round(n)))
    i += 1

df = pd.DataFrame(res, columns=['допустимая_ошибка', 'размер_выборки'])

print(df)
```
