```python
#task 3

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=",")

df.columns = [col.lower().replace(" ", "_") for col in df.columns]

a = df[["ингредиент_1", "брак"]].sort_values("ингредиент_1")

overall_mean = a["ингредиент_1"].mean()
overall_var = np.var(a["ингредиент_1"])

between_var = 0
within_var = 0

for group, group_df in a.groupby("брак"):
    group_mean = group_df["ингредиент_1"].mean()
    between_var += len(group_df) * (group_mean - overall_mean) ** 2
    within_var += (group_df["ингредиент_1"] - group_mean).pow(2).sum()

eta = np.sqrt(between_var / (between_var + within_var))

print(eta.round(4))
```

```python
#task 4

import pandas as pd
import numpy as np

df = pd.read_csv("data.csv", sep=",")

df.columns = [col.lower().replace(" ", "_") for col in df.columns]

df = df[(df["номер_конвейера"] == 1) & (df["номер_смены"] == 2)]

a = df[["вес", "брак"]].sort_values("вес")

overall_mean = a["вес"].mean()
overall_var = np.var(a["вес"])

between_var = 0
within_var = 0

for group, group_df in a.groupby("брак"):
    group_mean = group_df["вес"].mean()
    between_var += len(group_df) * (group_mean - overall_mean) ** 2
    within_var += (group_df["вес"] - group_mean).pow(2).sum()

eta = np.sqrt(between_var / (between_var + within_var))

print(eta.round(4))
```

```python
#task 5

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data.csv", sep=",")

df.columns = [col.lower().replace(" ", "_") for col in df.columns]

df = df[(df["ингредиент_1"] < 270) & (df["номер_смены"] == 2)]

a = df[["оператор_линии", "брак"]]

contingency = pd.crosstab(a["оператор_линии"], a["брак"])
chi2, p, dof, expected = chi2_contingency(contingency)

cramer = np.sqrt(chi2 / (a.shape[0] * (min(contingency.shape) - 1)))

print(cramer.round(4))
```

```python
#task 6

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data.csv", sep=",")

df = df.dropna()

df.columns = [col.lower().replace(" ", "_") for col in df.columns]

df["ингредиент_кат"] = [
    "менее_270" if x <= 270 else "более_270" for x in df["ингредиент_1"]
]

a = df[["ингредиент_кат", "брак"]]

contingency = pd.crosstab(a["ингредиент_кат"], a["брак"])
chi2, p, dof, expected = chi2_contingency(contingency)

cramer = np.sqrt(chi2 / (a.shape[0] * (min(contingency.shape) - 1)))

print(cramer.round(4))
```
